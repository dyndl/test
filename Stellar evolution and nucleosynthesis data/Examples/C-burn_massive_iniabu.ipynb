# C-burning shell in massive star
In this notebook we identify a C-burning shell in a massive star and extract abundances at a mass coordinate inside. These may then be used as initial abudnances for a PPN one-zone simulation.

FH 

```python
%pylab 
from nugridpy import nugridse as mp
from nugridpy import mesa as ms

# default data location (try mirror if default
# is not available)
data_dir="/tmp/nugrid_data"   

# mirror data locations (some mirrors 
# may not be available at some time)

# data_dir="/data/nugrid_cedar"  # may be default    
# data_dir="/data/nugrid_rpod4"    
# data_dir="/data/nugrid_vos"    


ms.set_nugrid_path(data_dir)
mp.set_nugrid_path(data_dir)
```

```python
s=ms.star_log(mass=20,Z=0.02)
```

```python
ifig=106;close(ifig);figure(ifig)
modstart=100
s.kip_cont(ifig=ifig,t0_model=modstart,modstart=modstart, modstop=10000,ixaxis='model_number', ylims=[0,8.])
```

```python
ifig=107;close(ifig);figure(ifig)
modstart=3700
s.kip_cont(ifig=ifig,t0_model=modstart,modstart=modstart, modstop=5400,ixaxis='age', ylims=[1.0,5.])
```

```python
pt=mp.se(mass=20,Z=0.02)
```

```python
species=['H-1','He-4','C-12','C-13','N-14','O-16']
ifig=108;close(ifig);figure(ifig)
pt.abu_profile(isos=species, ifig=ifig, fname=5180, logy=True)
ylim(-7,0)
xlim(1.0,5.)
```

```python
pt.abund_at_masscoordinate(5180,3.5,online=True)
```

```python
ifig=109;close(ifig);figure(ifig)
pt.plot('mass','temperature', fname=5180, logy=False)
#ylim(-7,0)
xlim(1.0,5.)
```

```python
float(pt.get('temperature_unit'))/1.e9
```
