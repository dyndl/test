# Figures of section 3 of the SYGMA paper

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sygma as s
import os
```

```python
## Parameter for standard best fit
iniZ_st=0.0126
imf_type_st='chabrier'
imf_bdys_st=[0.1,100] #based on paper p. 18 (and tests on p.19)
imf_yields_range_st=[0.8,100]  #0.8 based on calculations in paper p.19, paper uses 40Msun but then messses up plots, 100Msun good agreement.
sn1a_rate_st='exp' #based on paper .p. 23
exp_dtd_st=2e9  #based on paper p. 23
direct_norm_1a_st=0.016 #better fit #0.02, #based on paper p. 23 (a parameter, see also p22)
dt_st=7e6
transitionmass_st=8 #6.8 #chosen to match the AGB and massive star  intersection.
yield_interp_st='wiersma' # means we correct for initial abundance as in wiersma (independent if yields are net yields or not.)
netyields_on_st=True #tells the code we actual use tabula.ted net yields. 
wiersmamod_st=True
table_st='yield_tables/agb_and_massive_stars_portinari98_marigo01.txt'
exclude_masses_st=[] #needs to be empty else 60Msun exlcuded!, cannot justify NOT to use 6,7Msun
sn1a_table_st='yield_tables/sn1a_t03.txt'
iniabu_table_st='yield_tables/iniabu/iniab_solar_Wiersma.ppn'
#mgal = 1e4Msun; not set because default, but important for ejected masses
```

```python
s0w=s.sygma(iolevel=0,iniZ=iniZ_st,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp=yield_interp_st,
           netyields_on=netyields_on_st,
            wiersmamod=wiersmamod_st,
           table=table_st, \
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab2.0E-02GN93.ppn')


s0w_noyieldmod=s.sygma(iolevel=0,iniZ=iniZ_st,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp=yield_interp_st,
           netyields_on=netyields_on_st,
            wiersmamod=False,
           table=table_st, \
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab2.0E-02GN93.ppn')


s1w=s.sygma(iolevel=0,iniZ=0.004,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp=yield_interp_st,
           netyields_on=netyields_on_st,
            wiersmamod=wiersmamod_st,
           table=table_st, \
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab4.0E-03GN93.ppn')


s1w_noyieldmod=s.sygma(iolevel=0,iniZ=0.004,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp=yield_interp_st,
           netyields_on=netyields_on_st,
            wiersmamod=False,
           table=table_st, \
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab4.0E-03GN93.ppn')


s1w_0p001a=s.sygma(iolevel=0,iniZ=0.001,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp=yield_interp_st,
           netyields_on=netyields_on_st,
            wiersmamod=wiersmamod_st,
           table=table_st, \
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab1.0E-03GN93_alpha.ppn')
```

```python
s0n=s.sygma(iolevel=0,iniZ=0.02,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp='lin',
           netyields_on=False,
            wiersmamod=False,
           table='yield_tables/agb_and_massive_stars_nugrid_MESAonly_fryer12delay.txt',
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab2.0E-02GN93.ppn')

s1n=s.sygma(iolevel=0,iniZ=0.001,imf_type=imf_type_st,
           imf_bdys=imf_bdys_st, 
           imf_yields_range=[1,30],
           sn1a_rate=sn1a_rate_st, 
           exp_dtd=exp_dtd_st, 
           direct_norm_1a=direct_norm_1a_st, 
            dt=dt_st,
           transitionmass=transitionmass_st,
           yield_interp='lin',
           netyields_on=False,
            wiersmamod=False,
           table='yield_tables/agb_and_massive_stars_nugrid_MESAonly_fryer12delay.txt',
           exclude_masses=exclude_masses_st,
           sn1a_table=sn1a_table_st,iniabu_table='yield_tables/iniabu/iniab1.0E-03GN93_alpha.ppn')
```

## Fig. 4

```python
#%matplotlib nbagg
fign = 1
f, axarr = plt.subplots(3, 2,num=fign)
markevery=10
props = dict(boxstyle='square', facecolor='w', alpha=1)
for specie in ['Total','C','N','O','Si','Fe']:
#specie='Ba-132'
        if specie=='Total':
            plt.sca(axarr[0, 0])
        if specie=='C':
            plt.sca(axarr[0, 1])
        elif specie=='N':
            plt.sca(axarr[1, 0])
        elif specie=='O':
            plt.sca(axarr[1, 1])
        elif specie=='Si':
            plt.sca(axarr[2, 0])
        elif specie=='Fe':
            plt.sca(axarr[2, 1])
            
        if specie == 'Total':
            
            markevery=8            
            s0n.plot_totmasses(fig=fign,source='agb',markevery=markevery,color='b',marker='^')
            s0w.plot_totmasses(fig=fign,source='agb',markevery=markevery+5,color='r',marker='>')
            s0n.plot_totmasses(fig=fign,source='massive',markevery=markevery,color='b',marker='D')
            s0w.plot_totmasses(fig=fign,source='massive',markevery=markevery+5,color='r',marker='o')
            plt.plot([],[],marker='',linestyle='-.',color='k',label='AGBt',linewidth=3)
            plt.plot([],[],marker='',linestyle='-',color='k',label='Massive',linewidth=3)            
            ax=plt.gca()
            l2=ax.lines[-2:]
            plt.legend(l2,('AGB','Massive'),loc=6,prop={'size': 12}).set_visible(True)             
            
        else:  
            s0n.plot_mass(fig=fign,specie=specie,source='agb',norm='no',color='b',label=specie+' (AGB), Z=2e-2',shape='-.',marker='^',markevery=markevery)
            s0w.plot_mass(fig=fign,specie=specie,source='agb',norm='no',color='r',label=specie+' (AGB), Z=2e-2',shape='-.',marker='>',markevery=markevery+5)
            s0n.plot_mass(fig=fign,specie=specie,source='massive',norm='no',color='b',label=specie+' (massive), Z=2e-2',shape='-',marker='D',markevery=markevery)
            s0w.plot_mass(fig=fign,specie=specie,source='massive',norm='no',color='r',label=specie+' (massive), Z=2e-2',shape='-',marker='o',markevery=markevery+5)   
            plt.legend(loc=6,fontsize=12).set_visible(False)
        if specie in ['C',]:
            plt.ylim(1e-2,1e2)
        elif specie in ['Si']:
            plt.ylim(1e-1,2e1)
        elif specie in ['N']:
            plt.ylim(1e-1,2e1)
        elif specie in ['O']:
            plt.ylim(1e0,2e2)       
        elif specie in ['Fe']:
            plt.ylim(1e-1,1e1)
        ax=plt.gca()
        if specie=='Total':
            plt.plot([],[],marker='^',linestyle='-.',color='b',label='AGB')
            plt.plot([],[],marker='>',linestyle='-.',color='r',label='Massive')
            #plt.plot([],[],marker='+',linestyle=':',color='r',label='SNIa')
            plt.plot([],[],marker='D',linestyle='-',color='b',label='AGB')
            plt.plot([],[],marker='o',linestyle='-',color='r',label='Massive')
            l2=ax.lines[-4:]
            plt.legend(l2,('AGB, NuGrid','AGB, P98+M01','Massive, NuGrid','Massive, P98+M01'),loc=4,prop={'size': 14})
        ax=plt.gca()    
        if (specie=='N'):
            ax.set_ylabel(r'accumulated ejecta [M$_{\odot}$]',size=20)
        else:
            plt.ylabel('')
        if specie in ['Total','C','N','O']:
            ax.get_xaxis().set_ticklabels([])
            plt.xlabel('')    
        else:
            plt.xlabel('log-scaled age [yr]',size=20)
        plt.yscale('log')    
        ax=plt.gca()
        ax.text(0.1, 0.85, specie, transform=ax.transAxes, fontsize=16,verticalalignment='top', bbox=props)  
        if not specie =='Total':
            plt.ylim(1e-1,2e2)
#general settings
fig=plt.gcf()        
plt.subplots_adjust(hspace=0.1, bottom=0.125,wspace=0.15,left=0.1,right=0.9)
fig=plt.gcf()        
fig.set_size_inches(14,10,forward=True) #figsize=(6, 4)      
#fig.set_size_inches(12,10,forward=True) #figsize=(6, 4)  
fig.savefig('ssp_solar_z_wiersma_nugrid_yields.png', dpi=100) #save fgure
```
